# Ingest AIFS Initial Conditions

This notebook shows how to ingest ECMWF Open Data initial conditions for a given period into a Tigris-backed Icechunk repository.

## Prerequisites

- `AWS_ACCESS_KEY_ID` and `AWS_SECRET_ACCESS_KEY` exported in your environment.
- Optional: `AWS_REGION` (if your setup requires it).

```python
import os

import icechunk
import zarr

from aifs_modal_demo import ingest
```

```python
# Inclusive range, every 6 hours (00, 06, 12, 18 UTC).
start_date = "2026-02-16T00:00:00+00:00"
end_date = "2026-02-16T06:00:00+00:00"

storage_bucket = "aifs-modal"
initial_conditions_prefix = "aifs-initial-conditions"
initial_conditions_branch = "main"
```

```python
ingest.ingest(
    start_date=start_date,
    end_date=end_date,
    storage_bucket=storage_bucket,
    initial_conditions_prefix=initial_conditions_prefix,
    initial_conditions_branch=initial_conditions_branch,
)
```

```python
storage = icechunk.tigris_storage(
    bucket=storage_bucket,
    prefix=initial_conditions_prefix,
    region=os.getenv("AWS_REGION", None),
    access_key_id=os.environ["AWS_ACCESS_KEY_ID"],
    secret_access_key=os.environ["AWS_SECRET_ACCESS_KEY"],
)
repo = icechunk.Repository.open(storage)
session = repo.readonly_session(initial_conditions_branch)
root = zarr.open_group(session.store, mode="r", zarr_format=3)

groups = sorted(root.group_keys())
print("Top-level date groups:")
print(groups)
```
